```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere, plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
```

```python
qc = QuantumCircuit(1,1) #single Qubit circut (1,#) 1 for quantum and 1 for # Classical Register
qc.h(0)   # Single Qubit Gate
qc.measure(0,0) # Applying on zeroth qubit
qc.draw('mpl')
```

```python
## Running above circuit on Qunatum Simulator
from qiskit import Aer, execute, BasicAer
```

```python
!pip install -q qiskit-aer
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc,simulator, shots = 200).result()
counts = result.get_counts(qc)
plot_histogram(counts)
```

```python
BellC = QuantumCircuit(2,2)
BellC.h(0)
BellC.cx(0,1)
BellC.measure(0,0)
BellC.measure(1,1)
BellC.draw('mpl')
```

```python
BellC2 = QuantumCircuit(2)
BellC2.h(0)
BellC2.cx(0,1)
BellC2.draw('mpl')
```

```python
sv = Statevector.from_label('10')
plot_bloch_multivector(sv)
```
